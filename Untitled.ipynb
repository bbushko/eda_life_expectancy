# Exploratory Data Analysis
## Life Expectancy & Suicide Rate

Gaining a solid basis understanding of the data. Datasets from Kaggle: https://www.kaggle.com/datasets/marshuu/suicide-rate-and-life-expectancy?select=Life+expectancy.csv


Our main questions are:

- Does level of GDP correlates with life expectancy in a country?
- Does level of GDP correlates with suicide rate in a country?
- Who benefits more from high/low fertility rate in a country in terms of happiness score?
- What regions have more GDP/Happiness score?


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
import numpy as np
```

```python
life_expectancy = pd.read_csv("Life expectancy.csv")
suicide_rate = pd.read_csv("Suicide Rate.csv")
```

```python
life_expectancy.head()
```

```python
suicide_rate.head()
```

```python
df = pd.merge(life_expectancy, suicide_rate)
df.head()
```

```python
df.info()

# As we can see, our GDP per capita somehow is "object" type, while it should be float64.
# Let us see what is wrong.
```

```python
pd.to_numeric(df['GDP per capita'], downcast='float')
df.info()

# As we see, the problem is with the very first value in our df – "1,105". 
# Probably the problem is with number format: there are redundant commas.
```

```python
df['GDP per capita'] = df['GDP per capita'].str.replace(",", ".")
df = df.astype({"GDP per capita": float})

df.info()

# Problem is solved.
```

```python
df = df.astype({"GDP per capita": float})

df['GDP per capita'].describe()
```

__Code below tells us that distribution of GDP is far from normal, and is right skewed.__

```python
print(f"Skewness: {df['GDP per capita'].skew()}")
print(f"Kurtosis: {df['GDP per capita'].kurt()}")
```

```python
df['GDP per capita'].plot.hist(bins=20)

plt.axvline(np.mean(df['GDP per capita']), color='red', linestyle='dashed', linewidth=1)
    
plt.title("Distribution of GDP by value")
plt.xlabel("GDP, $")
plt.show()
```

```python
# A correlation matrix

seaborn.pairplot(df)
```

## From the pairplot we can assume that:

- Life Expectancy is highly positively correlated with Happiness Score, and negatively correlated with Fertility Rate. 

Therefore, people in happier countries probably live longer, than in less happy ones. Also, society in happier and welthier countries chooses to birth less. It is general trend to birth more in poor countries. Many characteristics of poverty can cause high fertility – high infant mortality, lack of education for women in particular, too little family income to invest in children, inequitable shares in national income, and the inaccessibility of family planning.

- We cannot assume that suicide rate is correlated with any variables in the dataset. 

Ergo, happiness and presence of children probably does not affect on a person's decision of whether or not to commit a suicide

- GDP correlates with Life Expectancy and Happiness Score.
